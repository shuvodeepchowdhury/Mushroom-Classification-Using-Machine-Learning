```python
for column in df.columns:
    df[column] = df[column].astype('category')
dummy = pd.get_dummies(df, drop_first=True, dtype=int)
dummy
```

```python
dum=pd.get_dummies(df, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=int)
dum
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
scaler = MinMaxScaler()
scaler1= StandardScaler()
dummy_scale = scaler.fit_transform(dum)
dummy_scale1 = scaler1.fit_transform(dum)
dummy_scaled_df = pd.DataFrame(dummy_scale, columns=dum.columns)
dummy_scaled_df1 =pd.DataFrame(dummy_scale1, columns=dum.columns)
dummy_scaled_df1
```

```python
for column in dummy_scaled_df1.columns:
    dummy_scaled_df1[column] = dummy_scaled_df1[column].astype('category')
dummy1 = pd.get_dummies(dummy_scaled_df1, drop_first=True, dtype=int)
dummy1
```

```python
scaler3 = StandardScaler()
dummy_standard_scaled = scaler3.fit_transform(dummy)
dummy_standard_scaled_df = pd.DataFrame(dummy_standard_scaled, columns=dummy.columns)
for column in dummy_standard_scaled_df.columns:
    dummy_standard_scaled_df[column] = dummy_standard_scaled_df[column].astype('category')
dummy_standard_scaled_df
```

```python
x = dummy_standard_scaled_df.drop("class_p",axis=1)
```

```python
x.head()
```

```python
y=dum.class_e
```

```python
y.head()
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = LogisticRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

print('Classification Report:')
print(classification_report(y_test, y_pred))

print('Confusion Matrix:')
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
y_pred_nb = nb_model.predict(x_test)
```

```python
print("Gaussian Naive Bayes Model Evaluation:")
print(f'Accuracy: {accuracy_score(y_test, y_pred_nb):.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred_nb))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_nb))
```

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(x_train, y_train)
y_pred_svm = svm_model.predict(x_test)

print("SVM Model Evaluation:")
print(f'Accuracy: {accuracy_score(y_test, y_pred_svm):.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred_svm))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_svm))
```

```python
import joblib
joblib.dump(nb_model, 'gaussian_nb_model.pkl')
joblib.dump(svm_model, 'svm_model.pkl')
joblib.dump(model,'logistic.pkl')
```

```python
loaded_logistic_model = joblib.load('logistic.pkl')
y_pred_loaded_logistic = loaded_logistic_model.predict(x_test)

print("\nLoaded logistic Model Evaluation:")
print(f'Accuracy: {accuracy_score(y_test, y_pred_loaded_logistic):.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred_loaded_logistic))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_loaded_logistic))

loaded_nb_model = joblib.load('gaussian_nb_model.pkl')
y_pred_loaded_nb = loaded_nb_model.predict(x_test)

print("Loaded Gaussian Naive Bayes Model Evaluation:")
print(f'Accuracy: {accuracy_score(y_test, y_pred_loaded_nb):.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred_loaded_nb))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_loaded_nb))

loaded_svm_model = joblib.load('svm_model.pkl')
y_pred_loaded_svm = loaded_svm_model.predict(x_test)

print("\nLoaded SVM Model Evaluation:")
print(f'Accuracy: {accuracy_score(y_test, y_pred_loaded_svm):.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred_loaded_svm))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_loaded_svm))
```
